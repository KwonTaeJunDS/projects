```python
import pandas as pd

data = pd.read_csv('광진구 공모전 데이터 전처리 완료.csv',encoding='utf-8')
data.head()
```

```python
# '일시' 컬럼을 datetime 객체로 변환
data['일시'] = pd.to_datetime(data['일시'])

# 월 정보 추출
data['월'] = data['일시'].dt.month

# 계절 정보를 추가하는 함수
def assign_season(month):
    if month in [3, 4, 5]:
        return '봄'
    elif month in [6, 7, 8]:
        return '여름'
    elif month in [9, 10, 11]:
        return '가을'
    else:
        return '겨울'

# 계절 컬럼 생성
data['계절'] = data['월'].apply(assign_season)

# 결과 확인
data[['일시', '월', '계절']].head()
```

```python
data = data.dropna()
data.shape
```

```python
from sklearn.model_selection import train_test_split

# 대상 변수인 '일강수량(mm)'와 특성들 선택
target = '일강수량(mm)'
features = data.columns.drop([target, '일시'])  # '일시'는 모델 입력으로 직접 사용하지 않고, 파생 변수인 '월'과 '계절' 사용

# 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)

# 분할된 데이터 크기 확인
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.preprocessing import OneHotEncoder

# OneHotEncoder 생성
encoder = OneHotEncoder(sparse=False)

# '계절' 컬럼 원-핫 인코딩
season_encoded = encoder.fit_transform(data[['계절']])

# 원-핫 인코딩 결과를 DataFrame으로 변환
season_encoded_df = pd.DataFrame(season_encoded, columns=encoder.get_feature_names_out(['계절']))

# 원본 데이터프레임에 인코딩된 데이터를 추가
data_encoded = data.drop('계절', axis=1).reset_index(drop=True)
data_encoded = pd.concat([data_encoded, season_encoded_df], axis=1)

# 다시 훈련 세트와 테스트 세트로 데이터 분할
features_encoded = data_encoded.columns.drop([target, '일시'])  # '일시'는 모델 입력으로 직접 사용하지 않음
X_train, X_test, y_train, y_test = train_test_split(data_encoded[features_encoded], data_encoded[target], test_size=0.2, random_state=42)

# 원-핫 인코딩 적용 후 다시 모델 훈련
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# 훈련 세트와 테스트 세트에서의 성능 평가
y_train_pred = rf_model.predict(X_train)
y_test_pred = rf_model.predict(X_test)

# 평균 제곱 오차 계산
train_mse = mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

train_mse, test_mse
```

```python
import numpy as np

# 80mm 초과 시 '위험'으로 분류하는 기능 추가
def classify_rainfall(rainfall):
    return "위험" if rainfall > 70 else "안전"

# 테스트 세트에 대한 예측 결과에 적용
y_test_classified = np.array([classify_rainfall(rain) for rain in y_test_pred])

# 예측 결과의 일부 확인
y_test_classified[:10]
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
import seaborn as sns

# 테스트 세트의 실제 일강수량과 예측된 일강수량 비교 시각화
plt.figure(figsize=(10, 6))
sns.scatterplot(x=y_test, y=y_test_pred, alpha=0.6)
plt.axhline(70, color='r', linestyle='--', label='위험 기준선 (70mm)')
plt.axvline(70, color='r', linestyle='--')
plt.xlabel('실제 일강수량 (mm)')
plt.ylabel('예측 일강수량 (mm)')
plt.title('실제 대비 예측 일강수량')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 랜덤포레스트 모델의 특성 중요도 가져오기
importances = rf_model.feature_importances_

# 특성 이름과 함께 중요도 출력
features = X_train.columns
importance_df = pd.DataFrame({'Feature': features, 'Importance': importances})
importance_df.sort_values(by='Importance', ascending=False, inplace=True)

print(importance_df)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 특성 중요도를 데이터 프레임으로 변환
feature_importances = rf_model.feature_importances_
features = X_train.columns
importance_df = pd.DataFrame({'Feature': features, 'Importance': feature_importances})

# 데이터를 중요도에 따라 내림차순 정렬
importance_df = importance_df.sort_values(by='Importance', ascending=False)

# 시각화
plt.figure(figsize=(10, 8))
sns.barplot(data=importance_df, x='Importance', y='Feature')
plt.title('Feature Importances')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```

```python
# 계절별 컬럼을 기반으로 원래의 계절 정보 복구
conditions = [
    (data_encoded['계절_봄'] == 1),
    (data_encoded['계절_여름'] == 1),
    (data_encoded['계절_가을'] == 1),
    (data_encoded['계절_겨울'] == 1)
]
choices = ['봄', '여름', '가을', '겨울']
data_encoded['계절'] = np.select(conditions, choices, default='알 수 없음')
# 계절별 평균 강수량 계산
seasonal_rainfall = data_encoded.groupby('계절')['Predicted_Rainfall'].mean()
# 계절별 평균 강수량 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x=seasonal_rainfall.index, y=seasonal_rainfall.values)
plt.title('Seasonal Average Predicted Rainfall')
plt.ylabel('Average Rainfall (mm)')
plt.xlabel('Season')
plt.show()
```

```python
monthly_rainfall = data_encoded.groupby('월')['Predicted_Rainfall'].mean()

# 월별 평균 강수량 시각화
plt.figure(figsize=(12, 6))
sns.barplot(x=monthly_rainfall.index, y=monthly_rainfall.values)
plt.title('Monthly Average Predicted Rainfall')
plt.ylabel('Average Rainfall (mm)')
plt.xlabel('Month')
plt.show()
```
