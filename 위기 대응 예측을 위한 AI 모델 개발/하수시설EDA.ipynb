```python
import pandas as pd

data1 = pd.read_csv('방수설비_20240409111220.csv')
print(data1.head(8))
```

```python
index_to_drop = data1.index[data1['소재지(시군구)별(1)'] == '부산광역시'][0]
data1 = data1.iloc[:index_to_drop]
print(data1)
```

```python
data1.info()
```

```python
data1.fillna('서울특별시',inplace=True)
data1.head(10)
```

```python
data1.drop([2,3],axis= 0,inplace=True)
```

```python
data1.reset_index(drop=True, inplace=True)
data1
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
df = data1[['소재지(시군구)별(2)','2020', '2021', '2022']]

# 인덱스 설정
df.set_index('소재지(시군구)별(2)', inplace=True)
df['2020'] = pd.to_numeric(df['2020'], errors='coerce')
df['2021'] = pd.to_numeric(df['2021'], errors='coerce')
df['2022'] = pd.to_numeric(df['2022'], errors='coerce')

# 데이터 시각화
df.plot(kind='bar', figsize=(10, 6))
plt.xlabel('지역')
plt.ylabel('소계')
plt.title('지역별 소계 변화')
plt.xticks(rotation=45)  # x 축 레이블 회전
plt.legend(title='연도')
plt.show()
```

```python
df_gwangjin = df.loc['광진구']

# 데이터 시각화
df_gwangjin.T.plot(kind='bar', figsize=(10, 6))
plt.xlabel('연도')
plt.ylabel('소계')
plt.title('광진구 소계 변화')
plt.xticks(rotation=0)  # x 축 레이블 회전
plt.show()
```

```python
data2 = pd.read_csv('서울특별시 광진구_수방시설물_20220118.csv', encoding='euc-kr')
data2.head(10)
```

```python
data2['위 치'] = data2['위 치'].str[:3]
data2
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False


# 위치를 기준으로 수방시설물을 그룹화하여 개수를 계산
location_counts = data2.groupby('위 치').size().reset_index(name='수방시설물 개수')

# 시각화
plt.figure(figsize=(10, 6))
sns.barplot(x='수방시설물 개수', y='위 치', data=location_counts, palette='viridis')
plt.title('위치별 수방시설물 개수')
plt.xlabel('수방시설물 개수')
plt.ylabel('위 치')
plt.show()
```

```python
data3 = pd.read_csv('하수도_20240409112142.csv')
data3.head(10)
```

```python
data3.drop('소재지(시군구)별(1)', axis = 1, inplace=True)
data3.head(10)
```

```python
data3.columns
```

```python

df = data3[['소재지(시군구)별(2)', '2022']]

# 필요 없는 행 삭제
df = df.drop([0, 1, 2])

# 인덱스 설정
df.set_index('소재지(시군구)별(2)', inplace=True)

# 데이터 타입 변환 (문자열에서 숫자로)
df['2022'] = pd.to_numeric(df['2022'], errors='coerce')

```

```python
df = data3[['소재지(시군구)별(2)', '2022.1']]

# 필요 없는 행 삭제
df = df.drop([0, 1, 2])

# 인덱스 설정
df.set_index('소재지(시군구)별(2)', inplace=True)

# 데이터 타입 변환 (문자열에서 숫자로)
df['2022.1'] = pd.to_numeric(df['2022.1'], errors='coerce')
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
plt.rcParams['axes.unicode_minus'] =False
plt.figure(figsize=(60, 60))
# 시각화
ax = df.plot(kind='bar', legend=False)
ax.patches[4].set_facecolor('orange')  # 광진구 막대 색깔 변경

# 시각화
ax.plot(kind='bar', legend=False)
plt.xlabel('장소')
plt.ylabel('하수시설 수')
plt.title('장소별 하수시설 수')
plt.show()
```
