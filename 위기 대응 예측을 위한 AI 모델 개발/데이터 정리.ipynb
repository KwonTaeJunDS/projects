```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'
import seaborn as sns
# 모든 변수에 대한 히스토그램 생성
num_vars = data.select_dtypes(include=['float64', 'int64']).columns
fig, axes = plt.subplots(nrows=(len(num_vars) + 1) // 2, ncols=2, figsize=(15, len(num_vars) * 2))
axes = axes.flatten()  # axes 배열을 1차원으로 변환

for i, var in enumerate(num_vars):
    sns.histplot(data[var], kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {var}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('')

# 빈 subplot 숨기기
for ax in axes[len(num_vars):]:
    ax.set_visible(False)

fig.tight_layout()
plt.show()
```

```python
# 모든 수치형 변수의 상관관계 계산
all_correlation_matrix = data.select_dtypes(include=['float64', 'int64']).corr()

# 상관관계 히트맵 시각화
plt.figure(figsize=(16, 12))
sns.heatmap(all_correlation_matrix, annot=False, cmap='coolwarm', fmt=".1f", linewidths=.5)
plt.title('Correlation Matrix of All Variables')
plt.show()
```

```python
# 절대값 0.7 이상의 상관관계를 가지는 변수 쌍 추출
strong_correlations = all_correlation_matrix.abs().unstack()
strong_correlations = strong_correlations[strong_correlations >= 0.7]
strong_correlations = strong_correlations[strong_correlations < 1]  # 자기 자신과의 상관관계 제외
strong_correlations = strong_correlations.drop_duplicates()  # 중복 제거

strong_correlations
```

```python


# 필요한 주요 변수 선택
key_variables = ['평균기온(°C)', '최저기온(°C)', '최고기온(°C)', '일강수량(mm)',\
                 '평균 풍속(m/s)', '최대 순간 풍속(m/s)', '평균 상대습도(%)', '평균 해면기압(hPa)']

# 이 변수들에 대한 분포 확인
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 20))
fig.tight_layout(pad=5.0)

for i, var in enumerate(key_variables):
    row, col = i // 2, i % 2
    sns.histplot(data[var], kde=True, ax=axes[row, col])
    axes[row, col].set_title(f'Distribution of {var}')

plt.show()

# 강수량과 다른 변수들과의 상관관계 계산 및 히트맵 시각화
correlation_matrix = data[key_variables].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation matrix of key variables')
plt.show()
```

```python
# 최고기온에서 누락된 값을 찾기
missing_temp = data[data['최고기온(°C)'].isna()]
missing_temp[['일시', '최고기온(°C)']]
```

```python
# 일시 데이터를 datetime 형식으로 변환
data['일시'] = pd.to_datetime(data['일시'])

# 가을 계절 정의 (9월 1일부터 11월 30일)
fall_temps = data[(data['일시'].dt.month >= 9) & (data['일시'].dt.month <= 11)]

# 가을 최고기온의 평균 계산
fall_avg_high_temp = fall_temps['최고기온(°C)'].mean()
fall_avg_high_temp
```

```python
# 누락된 최고기온 값 채우기
data['최고기온(°C)'].fillna(fall_avg_high_temp, inplace=True)

# 변경 후 데이터 확인
data.loc[1380, ['일시', '최고기온(°C)']]
```

```python
# 강수량 예측과 관련된 변수들에서 누락된 값 0으로 대체
rainfall_related_vars_with_nulls = ['일강수량(mm)', '강수 계속시간(hr)', '10분 최다 강수량(mm)', '1시간 최다강수량(mm)', '9-9강수(mm)']
data[rainfall_related_vars_with_nulls] = data[rainfall_related_vars_with_nulls].fillna(0)

# 변경 후 누락된 값 확인
data[rainfall_related_vars_with_nulls].isna().sum()
```

```python
# 강수량 예측과 관련된 변수들에서 누락된 값 0으로 대체
rainfall_related_vars_with_nulls = ['10분 최다강수량 시각(hhmi)','1시간 최다 강수량 시각(hhmi)']
data[rainfall_related_vars_with_nulls] = data[rainfall_related_vars_with_nulls].fillna(0)

# 변경 후 누락된 값 확인
data[rainfall_related_vars_with_nulls].isna().sum()
```

```python
# 안개 계속시간(hr)의 누락된 값 0으로 대체
data['안개 계속시간(hr)'].fillna(0, inplace=True)

# 변경 후 누락된 값 확인
data['안개 계속시간(hr)'].isna().sum()
```

```python
# 계절별로 그룹화하여 평균 풍속 계산
data['월'] = data['일시'].dt.month
seasons = {12: '겨울', 1: '겨울', 2: '겨울', 
           3: '봄', 4: '봄', 5: '봄', 
           6: '여름', 7: '여름', 8: '여름', 
           9: '가을', 10: '가을', 11: '가을'}

# 월을 계절로 매핑
data['계절'] = data['월'].map(seasons)

# 계절별 평균 풍속 계산
seasonal_avg_windspeed = data.groupby('계절')['평균 풍속(m/s)'].mean()
```

```python
for var in ['최대 순간 풍속(m/s)', '최대 풍속(m/s)', '풍정합(100m)', '최다풍향(16방위)']:
    seasonal_avg = data.groupby('계절')[var].mean()  # 계절별 평균 계산
    for season, avg in seasonal_avg.items():
        mask = (data['계절'] == season) & (data[var].isna())
        data.loc[mask, var] = avg  # 결측치 대체


```

```python
# 모든 풍속 관련 변수의 결측치 처리 후 상태 확인
data.isna().sum()
```

```python
# 2017년 10월(가을)의 평균 최소 상대습도 계산
fall_humidity_avg = data[(data['월'] >= 9) & (data['월'] <= 11)]['최소 상대습도(%)'].mean()

# 결측치 대체
data.loc[data['최소 상대습도(%)'].isna(), '최소 상대습도(%)'] = fall_humidity_avg

# 결측치 처리 후 확인
data.loc[1380, ['일시', '최소 상대습도(%)']]
```

```python
# 2017년 가을(9-11월)의 평균 최저 해면기압 계산
fall_avg_pressure = data[(data['월'] >= 9) & (data['월'] <= 11)]['최저 해면기압(hPa)'].mean()

# 결측치 대체
data.loc[data['최저 해면기압(hPa)'].isna(), '최저 해면기압(hPa)'] = fall_avg_pressure

# 결측치 처리 후 확인
data.loc[1380, ['일시', '최저 해면기압(hPa)']]
```
